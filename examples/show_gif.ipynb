```python
from matplotlib import pyplot, animation
import numpy as np
from open_atmos_jupyter_utils import show_gif, TemporaryFile

gif_file = TemporaryFile(suffix=".gif")
fig, ax = pyplot.subplots(); ax.set(xlim=(-1.2,1.2), ylim=(-1.2,0.2), aspect='equal')
t = np.sin(np.linspace(0, 4*np.pi, 100)) * 0.8
line, = ax.plot([], [], 'o-', lw=2)
animation.FuncAnimation(
    fig,
    lambda i: [line.set_data([0, np.sin(t[i])], [0,-np.cos(t[i])]), line][1:],
    frames=100,
    interval=40
).save(gif_file.basename, writer="pillow", fps=25)
show_gif(gif_file)
pyplot.close(fig)
```
